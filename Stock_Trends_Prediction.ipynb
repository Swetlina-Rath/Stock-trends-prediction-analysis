```python
from neuralprophet import NeuralProphet
import yfinance as yf
from datetime import datetime
```

```python
import matplotlib.pyplot as plt
```

# Loading Data

```python
company = 'GOOG'
start_date = '2004-01-01'
end_data = datetime.now()

stock = yf.download(company, start_date, end_data)
stock
```

```python
stock.to_csv('stock_google.csv')
```

```python
import pandas as pd
stock = pd.read_csv('stock_google.csv')
stock.info()
```

# Preprocessing

```python
stock['Date'] = pd.to_datetime(stock['Date'])
stock.info()
```

```python
data = stock[['Date','Close']]
data
```

```python
data.columns = ['ds','y']
data
```

# Visualization

```python
plt.plot(data['ds'], data['y'], label = 'actual', c = 'g')
plt.xlabel('years')
plt.ylabel('Closing Price')
plt.title('Google Closing Price')
plt.show()
```

```python
def plot_graph(values, column):
    plt.figure(figsize = (15,5))
    values.plot()
    plt.xlabel('years')
    plt.ylabel(column)
    plt.title(f'{column} Data of Google')
```

```python
stock.columns
```

```python
for column in stock.columns[1:len(stock.columns)]:
  plot_graph(stock[column],column)
```

# Train Model

```python
model = NeuralProphet()
model.fit(data)
```

# Evaluate Model

```python
future = model.make_future_dataframe(data, periods = 300)

forecast = model.predict(future)
```

```python
forecast
```

```python
forecast[['ds', 'yhat1']]
```

```python
actual_prediction = model.predict(data)
```

```python
plt.plot(actual_prediction['ds'], actual_prediction['yhat1'], label = 'actual_prediction', c = 'r')
plt.plot(forecast['ds'], forecast['yhat1'], label = 'future_prediction', c = 'b')
plt.plot(data['ds'], data['y'], label = 'actual', c = 'g')
plt.xlabel('years')
plt.ylabel('Price in USD')
plt.title('Predictions of Closing Data of Google')
plt.legend()
plt.show()
```

```python
plt.plot(forecast['ds'], forecast['yhat1'], label = 'future_prediction', c = 'b')
plt.xlabel('years')
plt.ylabel('Price in USD')
plt.title('Future Trend Prediction of Google Closing Price')
plt.show()
```

```python
model.plot_components(forecast)
```
